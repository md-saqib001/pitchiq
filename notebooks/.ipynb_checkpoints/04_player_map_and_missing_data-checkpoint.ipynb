# PitchIQ - Day 4: Final Player Map & Missing Data Investigation

Finalizing the manual player canonical mapping and investigating data completeness (missing fielders, missing innings, truncated overs) to build a quality scorecard.

```python
import os
import json
import pandas as pd

data_dir = '../data/data'
json_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
```

### PART A - Final Player Canonical Map
Using the 3 manually confirmed duplicates and ignoring the 25 false positives.

```python
# 1. Extract all raw players first
all_players = set()
for file in json_files:
    with open(os.path.join(data_dir, file), 'r') as f:
        try:
            data = json.load(f)
            for inn in data.get('innings', []):
                for over in inn.get('overs', []):
                    for d in over.get('deliveries', []):
                        if 'batter' in d: all_players.add(d['batter'])
                        if 'bowler' in d: all_players.add(d['bowler'])
                        if 'non_striker' in d: all_players.add(d['non_striker'])
        except:
            pass

# 2. Build the Identity Map (everyone maps to themselves by default)
PLAYER_CANONICAL_MAP = {p: p for p in all_players}

# 3. Apply the 3 CONFIRMED overrides from manual review
PLAYER_CANONICAL_MAP['Arshad Khan (2)'] = 'Arshad Khan'
PLAYER_CANONICAL_MAP['Ravi Bishnoi'] = 'R Bishnoi' # Usually shorter name is canonical in Cricsheet, but you can swap these
PLAYER_CANONICAL_MAP['NA Saini'] = 'N Saini'

print(f"Raw Unique Players: {len(all_players)}")
canonical_count = len(set(PLAYER_CANONICAL_MAP.values()))
print(f"Cleaned Unique Players: {canonical_count}")
print(f"Total drops: {len(all_players) - canonical_count} (exactly matches the 3 confirmed groups!)")

# 4. Save mapping for ETL script
os.makedirs('../data', exist_ok=True)
with open('../data/player_canonical_map.json', 'w') as f:
    json.dump(PLAYER_CANONICAL_MAP, f, indent=2)
print("Saved PLAYER_CANONICAL_MAP to data/player_canonical_map.json")
```

### PART B - Missing Data Investigation
Checking for truncated overs, missing fielders on catches, and abandoned matches.

```python
missing_fielder_by_season = {}
match_scorecards = []

for file in json_files:
    with open(os.path.join(data_dir, file), 'r') as f:
        try:
            data = json.load(f)
            info = data.get('info', {})
            season = str(info.get('season'))
            if season not in missing_fielder_by_season:
                missing_fielder_by_season[season] = {"caught_total": 0, "caught_missing_fielder": 0}
            
            innings = data.get('innings', [])
            has_innings1 = len(innings) > 0
            has_innings2 = len(innings) > 1
            
            match_missing_fielders = 0
            match_over_mismatches = 0
            total_runs_scored = 0
            
            for inn in innings:
                for over in inn.get('overs', []):
                    legal_deliveries = 0
                    for d in over.get('deliveries', []):
                        # Add to total runs to check against target
                        runs = d.get('runs', {})
                        total_runs_scored += runs.get('total', 0)
                        
                        extras = d.get('extras', {})
                        # wides and noballs don't count as legal deliveries
                        if not ('wides' in extras or 'noballs' in extras):
                            legal_deliveries += 1
                        
                        if 'wickets' in d:
                            for w in d['wickets']:
                                if w['kind'] == 'caught':
                                    missing_fielder_by_season[season]['caught_total'] += 1
                                    # Check if fielders array is completely absent or empty
                                    if 'fielders' not in w or len(w['fielders']) == 0:
                                        match_missing_fielders += 1
                                        missing_fielder_by_season[season]['caught_missing_fielder'] += 1
                    
                    # Flag over if it ended with fewer than 6 legal deliveries (excluding rain/abandoned end cases)
                    if legal_deliveries < 6:
                        match_over_mismatches += 1
            
            match_scorecards.append({
                "match_id": file.split('.')[0],
                "season": season,
                "has_complete_innings1": has_innings1,
                "has_complete_innings2": has_innings2,
                "missing_fielder_count": match_missing_fielders,
                "over_count_mismatches": match_over_mismatches,
                "total_runs_in_balls": total_runs_scored
            })
        except:
            pass

df_scorecard = pd.DataFrame(match_scorecards)
print(f"Evaluated {len(df_scorecard)} matches into scorecard.")
```

### Fielder Missing Data Report by Season
Checking specifically if pre-2015 is lacking data.

```python
print("Missing Fielder on Caught Dismissal (by Season):\n")
for s, stats in sorted(missing_fielder_by_season.items(), key=lambda x: str(x[0])):
    pct = 0
    if stats['caught_total'] > 0:
        pct = stats['caught_missing_fielder'] / stats['caught_total'] * 100
    print(f"Season {s}: {stats['caught_missing_fielder']}/{stats['caught_total']} ({pct:.1f}%) missing fielders")
```

### Abandoned / Missing Innings Report
Which matches have no innings 2 data entirely?

```python
no_innings2 = df_scorecard[df_scorecard['has_complete_innings2'] == False]
print(f"Matches missing Innings 2 completely: {len(no_innings2)}")
print(no_innings2[['match_id', 'season', 'has_complete_innings1', 'has_complete_innings2']])
```

### Over Mismatch Report
How many overs ended prematurely with < 6 legal deliveries?

```python
total_mismatches = df_scorecard['over_count_mismatches'].sum()
print(f"Total overs across all matches ending with < 6 legal deliveries: {total_mismatches}")

# Show matches with the most truncated overs (often rain affected matches!)
print("\nTop 10 matches with the most truncated overs:")
print(df_scorecard.sort_values('over_count_mismatches', ascending=False)[['match_id', 'season', 'over_count_mismatches']].head(10))
```
