# PitchIQ - Day 5: Null Handling & Strategy

Finalizing the data rules for missing fielders, abandoned matches, truncated overs, and legal delivery definitions.

### 1. Missing Fielder on Caught Dismissals
**Decision:** Do nothing! As verified in Day 4, Cricsheet has retroactively fixed all old missing fielders. There are 0 cases of a `caught` dismissal missing a `fielders` array across 1226 matches. If one does appear in future data updates, we will store `NULL` in the database to maintain data integrity, and handle it gracefully as "Unknown" in the UI.

### 2. Matches Missing Innings 2 (Abandoned)
**Decision:** Keep the matches, but flag them. We want to preserve the Innings 1 deliveries for player stats, but exclude these matches from any query that requires a match result (like average in a chase).

```python
def get_excluded_match_ids():
    """
    Returns a list of match IDs that were abandoned or had no result,
    specifically those missing an entire innings 2.
    """
    return [
        "1359519", 
        "1473492", 
        "1473495", 
        "1527685", 
        "501265", 
        "829763"
    ]

print(f"Flagged {len(get_excluded_match_ids())} matches for exclusion from result-based aggregations.")
```

### 3. Investigating Over-Count Mismatches (<6 balls)
We found 730 overs that ended prematurely. Let's inspect Match 829757, which had 2 truncated overs.

```python
import json

with open('../data/data/829757.json', 'r') as f:
    data = json.load(f)

for inn in data.get('innings', []):
    for over in inn.get('overs', []):
        legal_count = 0
        for d in over['deliveries']:
            extras = d.get('extras', {})
            if not ('wides' in extras or 'noballs' in extras):
                legal_count += 1
        
        if legal_count < 6:
            print(f"Innings: {inn['team']}, Over {over['over']}: {legal_count} legal deliveries.")
            for d in over['deliveries']:
                print(f"  {d['batter']} to {d['bowler']} | Runs: {d.get('runs', {}).get('total')} | Wicket: {d.get('wickets', [{}])[0].get('kind', 'None')}")
            print("-" * 40)
```

**Conclusion:** This is **NOT A BUG**. The first truncated over ended after 2 balls because the batting team was bowled out (10th wicket taken). The second truncated over ended after 3 balls because the chasing team hit the winning runs! We will NOT discard or "fix" these overs.

### 4. Legal Delivery Logic
Wides and no-balls do not count towards the 6 balls of an over, and they do not count as a "ball faced" for a batter's strike rate.

```python
def is_legal_delivery(delivery_dict):
    """
    Returns True if the delivery is legal (not a wide or no-ball).
    """
    extras = delivery_dict.get('extras', {})
    if 'wides' in extras or 'noballs' in extras:
        return False
    return True

# Manual Verification Test
test_over = data['innings'][0]['overs'][0]['deliveries'] # 1st over of Match 829757
legal_count = sum(1 for d in test_over if is_legal_delivery(d))
print(f"Total raw deliveries in test over: {len(test_over)}")
print(f"Calculated legal deliveries: {legal_count}")
if legal_count == 6:
    print("SUCCESS: Logic correctly filters to exactly 6 legal deliveries!")
```
