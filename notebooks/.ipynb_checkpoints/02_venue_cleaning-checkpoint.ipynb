# PitchIQ - Day 2: Venue Cleaning

Collapsing raw stadium string variants into canonical names and enriching them with geographic metadata.

```python
import os
import json
import pandas as pd

# 1. Load the raw data to get all the venues
data_dir = '../data/data'
records = []
for file in os.listdir(data_dir):
    if file.endswith('.json'):
        with open(os.path.join(data_dir, file), 'r') as f:
            try:
                match_data = json.load(f)
                info = match_data.get('info', {})
                records.append({
                    'filename': file,
                    'season': info.get('season'),
                    'venue': info.get('venue'),
                    'city': info.get('city')
                })
            except:
                pass

df = pd.DataFrame(records)
print(f"Loaded {len(df)} matches.")
```

### c. Check for Unmapped Venues
Ensuring we didn't miss any strings.

```python
unmapped = df[df['venue'].notnull() & df['venue_canonical'].isnull()]['venue'].unique()

if len(unmapped) > 0:
    print(f"WARNING: Found {len(unmapped)} unmapped venue strings:")
    for v in unmapped:
        print(f" - {v}")
else:
    print("SUCCESS: 100% of venue strings successfully mapped to canonical names!")
```

### d. Final Venue Counts

```python
raw_count = df['venue'].nunique()
canonical_count = df['venue_canonical'].nunique()

print("--- VENUE CLEANING RESULTS ---")
print(f"Raw String Variants: {raw_count}")
print(f"Actual Unique Stadiums: {canonical_count}")
print(f"Reduced by: {raw_count - canonical_count} redundant names")
```

### e. Save the cleaned mapping

```python
os.makedirs('../data', exist_ok=True)
map_path = '../data/venue_canonical_map.json'

with open(map_path, 'w') as f:
    json.dump(VENUE_CANONICAL_MAP, f, indent=2)

print(f"Saved canonical mapping to {map_path}")
```
