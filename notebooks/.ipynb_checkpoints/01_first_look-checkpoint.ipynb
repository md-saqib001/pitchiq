# PitchIQ - Day 1: Data Exploration

First look at the raw IPL ball-by-ball JSON data from Cricsheet.

```python
import os
import json
import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming the data is downloaded to a folder named 'data/ipl_json' at the project root
data_dir = '../data/data/ipl_json'

# Optional: Create the directory if it doesn't exist so the script doesn't fail immediately
os.makedirs(data_dir, exist_ok=True)
```

### a. List all JSON files in the data folder, print total count

```python
json_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
print(f"Total JSON files found: {len(json_files)}")
```

### b. Load ONE match JSON file, pretty-print its full structure

```python
if json_files:
    first_file = os.path.join(data_dir, json_files[0])
    with open(first_file, 'r') as f:
        sample_match = json.load(f)
    print(json.dumps(sample_match, indent=2))
else:
    print("No JSON files found. Please download the data first.")
```

### c. Print just the "info" section across 5 different random matches

```python
if json_files:
    sample_files = random.sample(json_files, min(5, len(json_files)))
    for file in sample_files:
        with open(os.path.join(data_dir, file), 'r') as f:
            match_data = json.load(f)
        info = match_data.get('info', {})
        
        print(f"--- File: {file} ---")
        print(f"City: {info.get('city')}")
        print(f"Venue: {info.get('venue')}")
        print(f"Season: {info.get('season')}")
        print(f"Teams: {info.get('teams')}")
        print(f"Toss: {info.get('toss')}")
        print(f"Outcome: {info.get('outcome')}")
        print()
```

### d. Build a pandas DataFrame with ONE row per match

```python
records = []
for file in json_files:
    with open(os.path.join(data_dir, file), 'r') as f:
        match_data = json.load(f)
    info = match_data.get('info', {})
    
    teams = info.get('teams', [])
    team1 = teams[0] if len(teams) > 0 else None
    team2 = teams[1] if len(teams) > 1 else None
    
    outcome = info.get('outcome', {})
    winner = outcome.get('winner')
    
    records.append({
        'filename': file,
        'season': info.get('season'),
        'venue': info.get('venue'),
        'city': info.get('city'),
        'team1': team1,
        'team2': team2,
        'winner': winner
    })

df = pd.DataFrame(records)
df.head()
```

### e. Print: df['season'].unique() sorted

```python
# Convert to string because seasons might be a mix of ints (2023) and strings ('2020/21')
unique_seasons = sorted([str(s) for s in df['season'].dropna().unique()])
print("Unique Seasons:")
for s in unique_seasons:
    print(s)
```

### f. Print: df['venue'].unique() sorted alphabetically

```python
unique_venues = sorted([str(v) for v in df['venue'].dropna().unique()])
print("Unique Venues:")
for v in unique_venues:
    print(v)
```

### g. Print: df['venue'].nunique() vs a rough estimate

```python
venue_count = df['venue'].nunique()
print(f"Total Unique Venue Strings: {venue_count}")
print("Estimated actual distinct stadiums in IPL history: ~35-55")

if venue_count > 55:
    diff = venue_count - 55
    print(f"\nWe have roughly {diff} redundant/inconsistent venue names to clean up!")
```
