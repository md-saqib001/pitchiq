# PitchIQ - Day 3: Season & Player Cleaning

Standardizing season strings to calendar years/season numbers, and using fuzzy matching to find player name inconsistencies across deliveries.

```python
import os
import json
import pandas as pd

# 1. Load the raw data for match metadata
data_dir = '../data/data'
json_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]

records = []
for file in json_files:
    with open(os.path.join(data_dir, file), 'r') as f:
        try:
            match_data = json.load(f)
            info = match_data.get('info', {})
            records.append({
                'filename': file,
                'season': info.get('season')
            })
        except:
            pass

df = pd.DataFrame(records)
print(f"Loaded {len(df)} matches.")
```

### PART A - Seasons
Mapping string variants like '2007/08' to actual calendar years using the physical match dates.

```python
SEASON_CANONICAL_MAP = {
    "2007/08": 2008,
    "2009": 2009,
    "2009/10": 2010,
    "2011": 2011,
    "2012": 2012,
    "2013": 2013,
    "2014": 2014,
    "2015": 2015,
    "2016": 2016,
    "2017": 2017,
    "2018": 2018,
    "2019": 2019,
    "2020/21": 2020,
    "2021": 2021,
    "2022": 2022,
    "2023": 2023,
    "2024": 2024,
    "2025": 2025,
    "2026": 2026
}

df['season_year'] = df['season'].map(SEASON_CANONICAL_MAP)
df['season_number'] = df['season_year'] - 2007

print(f"Missing mapped seasons: {df['season_year'].isna().sum()}")
print("\n--- CANONICAL SEASON MAPPING ---")
print(df[['season', 'season_year', 'season_number']].drop_duplicates().sort_values('season_number').to_string(index=False))
```

### PART B - Player Names
Extracting every unique batter, bowler, and non-striker to find inconsistencies.

```python
players = set()

for file in json_files:
    with open(os.path.join(data_dir, file), 'r') as f:
        try:
            data = json.load(f)
            innings = data.get('innings', [])
            for inn in innings:
                for over in inn.get('overs', []):
                    for d in over.get('deliveries', []):
                        if 'batter' in d: players.add(d['batter'])
                        if 'bowler' in d: players.add(d['bowler'])
                        if 'non_striker' in d: players.add(d['non_striker'])
        except:
            pass

players = sorted(list(players))
print(f"Total unique player names extracted: {len(players)}")
```

Using `thefuzz` to perform fuzzy string matching to detect likely duplicates (e.g. 'M S Dhoni' vs 'MS Dhoni').

```python
from thefuzz import fuzz

groups = []
matched = set()

for p1 in players:
    if p1 in matched: continue
    group = [p1]
    for p2 in players:
        if p1 != p2 and p2 not in matched:
            # token_sort_ratio ignores word order, very useful for names
            if fuzz.token_sort_ratio(p1, p2) > 85:
                group.append(p2)
    
    if len(group) > 1:
        groups.append(group)
        for p in group:
            matched.add(p)

# Sort by largest groups first (most likely to have major inconsistencies)
groups.sort(key=lambda x: len(x), reverse=True)

print(f"Found {len(groups)} potential duplicate groups.\n")
for i, g in enumerate(groups[:30]): # Show top 30
    print(f"Group {i+1}: {g}")
```
