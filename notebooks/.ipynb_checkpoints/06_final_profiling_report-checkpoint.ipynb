# PitchIQ - Day 6: Final Profiling Report & Export

Applying all cleaning steps end-to-end to generate a comprehensive data quality scorecard, exporting the flattened intermediate data, and spot-checking our results.

```python
import os
import json
import pandas as pd

# 1. Load configuration and rules
with open('../data/venue_canonical_map.json', 'r') as f:
    VENUE_MAP = json.load(f)

with open('../data/player_canonical_map.json', 'r') as f:
    PLAYER_MAP = json.load(f)

SEASON_MAP = {
    "2007/08": 2008, "2009": 2009, "2009/10": 2010, "2011": 2011, "2012": 2012,
    "2013": 2013, "2014": 2014, "2015": 2015, "2016": 2016, "2017": 2017,
    "2018": 2018, "2019": 2019, "2020/21": 2020, "2021": 2021, "2022": 2022,
    "2023": 2023, "2024": 2024, "2025": 2025, "2026": 2026
}

# Found in Day 5
EXCLUDED_MATCHES = ["1359519", "1473492", "1473495", "1527685", "501265", "829763"]

data_dir = '../data/data'
json_files = [f for f in os.listdir(data_dir) if f.endswith('.json')]
```

### End-to-End Extraction and Flattening

```python
deliveries_flat = []

stats = {
    "total_matches": 0, "total_excluded": 0, "total_deliveries": 0,
    "null_fielders": 0, "legal_deliveries": 0, "wides": 0, "noballs": 0
}

raw_venues, clean_venues = set(), set()
raw_players, clean_players = set(), set()

for file in json_files:
    match_id = file.split('.')[0]
    
    with open(os.path.join(data_dir, file), 'r') as f:
        try:
            data = json.load(f)
            stats["total_matches"] += 1
            if match_id in EXCLUDED_MATCHES: stats["total_excluded"] += 1
            
            info = data.get('info', {})
            raw_season = str(info.get('season'))
            season_year = SEASON_MAP.get(raw_season)
            season_number = season_year - 2007 if season_year else None
            
            raw_venue = str(info.get('venue'))
            clean_venue = VENUE_MAP.get(raw_venue, raw_venue)
            raw_venues.add(raw_venue)
            clean_venues.add(clean_venue)
            
            for inn in data.get('innings', []):
                team = inn.get('team')
                for over in inn.get('overs', []):
                    over_num = over.get('over')
                    for d_num, d in enumerate(over.get('deliveries', [])):
                        stats["total_deliveries"] += 1
                        
                        raw_batter = d.get('batter')
                        raw_bowler = d.get('bowler')
                        raw_non_striker = d.get('non_striker')
                        
                        raw_players.update([raw_batter, raw_bowler, raw_non_striker])
                        
                        clean_batter = PLAYER_MAP.get(raw_batter, raw_batter)
                        clean_bowler = PLAYER_MAP.get(raw_bowler, raw_bowler)
                        clean_non_striker = PLAYER_MAP.get(raw_non_striker, raw_non_striker)
                        clean_players.update([clean_batter, clean_bowler, clean_non_striker])
                        
                        extras = d.get('extras', {})
                        is_wide = 'wides' in extras
                        is_noball = 'noballs' in extras
                        is_legal = not (is_wide or is_noball)
                        
                        if is_legal: stats["legal_deliveries"] += 1
                        if is_wide: stats["wides"] += 1
                        if is_noball: stats["noballs"] += 1
                        
                        dismissal_kind = None
                        player_out = None
                        fielder = None
                        
                        if 'wickets' in d:
                            w = d['wickets'][0]
                            dismissal_kind = w.get('kind')
                            player_out_raw = w.get('player_out')
                            player_out = PLAYER_MAP.get(player_out_raw, player_out_raw)
                            
                            if dismissal_kind == 'caught':
                                if 'fielders' in w and len(w['fielders']) > 0:
                                    fielder_raw = w['fielders'][0].get('name')
                                    fielder = PLAYER_MAP.get(fielder_raw, fielder_raw)
                                else:
                                    stats["null_fielders"] += 1
                        
                        deliveries_flat.append({
                            "match_id": match_id,
                            "season_year": season_year,
                            "season_number": season_number,
                            "venue": clean_venue,
                            "innings": team,
                            "over": over_num,
                            "delivery_num": d_num + 1,
                            "batter": clean_batter,
                            "bowler": clean_bowler,
                            "non_striker": clean_non_striker,
                            "batter_runs": d.get('runs', {}).get('batter', 0),
                            "extra_runs": d.get('runs', {}).get('extras', 0),
                            "total_runs": d.get('runs', {}).get('total', 0),
                            "is_legal_delivery": is_legal,
                            "is_wide": is_wide,
                            "is_noball": is_noball,
                            "dismissal_kind": dismissal_kind,
                            "player_out": player_out,
                            "fielder": fielder,
                            "is_abandoned": match_id in EXCLUDED_MATCHES
                        })
        except:
            pass

df = pd.DataFrame(deliveries_flat)
```

### Generate Parquet File (The Cleaned Checkpoint)
We use `.parquet` over `.csv` because it is heavily compressed, stores column data types natively (no converting "True" from a string to a boolean), and is massively faster to query for analytical workloads.

```python
df.to_parquet('../data/cleaned_deliveries.parquet', index=False)
print("Saved data/cleaned_deliveries.parquet")
```

### Final Summary Report

```python
print("--- FINAL SUMMARY REPORT ---")
print(f"Total matches processed: {stats['total_matches']}")
print(f"Total matches excluded (abandoned/no result): {stats['total_excluded']}")
print(f"Total deliveries processed: {stats['total_deliveries']}")
print(f"Total unique canonical venues: {len(clean_venues)} (down from {len(raw_venues)} raw strings)")
print(f"Total unique canonical players: {len(clean_players)} (down from {len(raw_players)} raw strings)")
print(f"Season range covered: 2008 to 2026 (Season 1 to {len(SEASON_MAP)})")
print(f"Total null fielder entries: {stats['null_fielders']}")
print(f"Legal deliveries vs extras breakdown: {stats['legal_deliveries']} legal, {stats['wides']} wides, {stats['noballs']} no-balls")
```

### The 3-Match Sanity Spot Check
Let's calculate the final score of 3 random matches and compare them to the public reality (e.g., ESPNCricinfo) to ensure parsing logic is fully accurate.

1. **335982**: First IPL Match Ever (RCB vs KKR in 2008, KKR won thanks to McCullum's 158*)
2. **1082613**: IPL 2017 Final (MI vs RPS, MI won by 1 run)
3. **1304116**: IPL 2022 Final (GT vs RR, GT won)

```python
def spot_check(match_id):
    match_df = df[df['match_id'] == match_id]
    print(f"\nMATCH ID {match_id} | Venue: {match_df['venue'].iloc[0]} | Season: {match_df['season_year'].iloc[0]}")
    
    for inn in match_df['innings'].unique():
        inn_df = match_df[match_df['innings'] == inn]
        total_runs = inn_df['total_runs'].sum()
        wickets = inn_df['dismissal_kind'].notna().sum()
        print(f"  {inn} -> {total_runs}/{wickets}")
        
spot_check("335982")
spot_check("1082613")
spot_check("1304116")
```
